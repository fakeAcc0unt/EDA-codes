## UPI Transactions Dataset (Sampling & Sampling Distribution)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('MyTransaction.csv')
```

```python
df
```

```python
samples = pd.DataFrame({
    's1': df['Balance'].sample(50).values,
    's2': df['Balance'].sample(50).values,
    's3': df['Balance'].sample(50).values,
    's4': df['Balance'].sample(50).values,
    's5': df['Balance'].sample(50).values,
    's6': df['Balance'].sample(50).values,
})
```

```python
samples
```

```python
samples['Sample Mean'] = samples.mean(axis=1)
```

```python
samples
```

```python
sns.distplot(samples['Sample Mean'].iloc[:20], color = 'green', hist_kws={'edgecolor' : 'black'}, bins = 10)
```

```python
sns.distplot(samples['Sample Mean'], color = 'blue', hist_kws={'edgecolor' : 'black'}, bins = 10)
```
